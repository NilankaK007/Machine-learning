```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['BMI'], data['Outcome'], color='blue', alpha=0.6)
plt.title('Scatter Plot: BMI vs Charges')
plt.xlabel('BMI (Independent Variable)')
plt.ylabel('Charges (Dependent Variable)')
plt.grid(True)
plt.show()
```

```python
X=data[['BMI']]
Y=data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
plt.figure(figsize=(10, 6))

# Plot the Observed Values (Actual Data)
plt.scatter(X, Y, color='blue', alpha=0.5, label='Observed Values')

# Plot the Predicted Values (The Regression Line)
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Predicted Values (Model)')

plt.title('Model Estimation: Predicted vs Observed Values')
plt.xlabel('BMI (Independent Variable)')
plt.ylabel('Charges (Dependent Variable)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
print('Mean square error :',mse)
```

```python
r2=r2_score(y_test,y_pred)
print('R2 score:',r2)
```
