```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("advertising.csv")
dataset.head()
```

```python
#dropping unnecessary columns
dataset.drop(columns=['Radio','Newspaper'],inplace=True)
dataset.head()
```

```python
x=dataset[['TV']]
y=dataset['Sales']
#print(type(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
#print(x_train)
#print(x_test)
```

```python
from sklearn.linear_model import LinearRegression
slr=LinearRegression()
slr.fit(x_train,y_train)
#print(type(x_train))
```

```python
print("Intercept: ",slr.intercept_)
print("Slope: ",slr.coef_)
```

```python
y_pred_slr=slr.predict(x_test)
print("Prediction for test set:{}".format(y_pred_slr))
#print(type(y_pred_slr))
```

```python
slr_diff=pd.DataFrame({'Actual value': y_test,'Predicted value' : y_pred_slr})
slr_diff.head()
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred_slr,'Maroon')
plt.show()
```

```python
from sklearn import metrics
meanAberr=metrics.mean_absolute_error(y_test,y_pred_slr)
meanSqerr=metrics.mean_squared_error(y_test,y_pred_slr)
rootMeansqerr=np.sqrt(metrics.mean_squared_error(y_test,y_pred_slr))
print("R squared: {:2f}".format(slr.score(x,y)*100))
print("Mean absolue error: ",meanAberr)
print("Mean squared error: ",meanSqerr)
print("Root Mean squared error: ",rootMeansqerr)
```

```python
dataset=pd.read_csv("advertising.csv")
x=dataset[['TV','Radio','Newspaper']]
y=dataset['Sales']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
print("Intercept: ",mlr.intercept_)
print("Coefficients: ")
list(zip(x,mlr.coef_))
```

```python
y_pred_mlr=mlr.predict(x_test)
print("Prediction for test set:{}".format(y_pred_mlr))
```

```python
mlr_diff=pd.DataFrame({'Actual value': y_test,'Predicted value' : y_pred_mlr})
mlr_diff.head()
```

```python
plt.scatter(x_test["Radio"],y_test)
#plt.plot(x_test["Radio"],y_pred_mlr,'Red')
plt.show()
```

```python
from sklearn import metrics
meanAberr=metrics.mean_absolute_error(y_test,y_pred_mlr)
meanSqerr=metrics.mean_squared_error(y_test,y_pred_mlr)
rootMeansqerr=np.sqrt(metrics.mean_squared_error(y_test,y_pred_mlr))
print("R squared: {:2f}".format(mlr.score(x,y)*100))
print("Mean absolue error: ",meanAberr)
print("Mean squared error: ",meanSqerr)
print("Root Mean squared error: ",rootMeansqerr)
```
